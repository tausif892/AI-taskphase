```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine_dataset=load_wine( return_X_y=False, as_frame=True)
```

```python
import pandas  as pd
```

```python
wine=pd.DataFrame(wine_dataset.data,columns=wine_dataset.feature_names)
```

```python
wine['label']=wine_dataset.target
```

```python
import seaborn as sns
```

```python
sns.heatmap(wine.corr(),annot=True,annot_kws={'size':6})
```

```python
label=wine['label']
```

```python
depend=wine.drop(columns="label")
```

```python
from sklearn.model_selection import train_test_split
```

```python
label_train,label_test,depend_train,depend_test=train_test_split(label,depend,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(depend_train,label_train)
```

```python
y_pred=lr.predict(depend_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
acc_score=mean_absolute_error(label_test,y_pred)
```
