```python
import seaborn as sns

iris=sns.load_dataset('iris')
iris
```

```python
iris=iris[['petal_length','petal_width']]
iris
```

```python
X=iris[['petal_length']]
X
```

```python
Y=iris[['petal_width']]
Y
```

```python
import numpy as np
X=np.array(X)
Y=np.array(Y)
```

```python
X=np.array(X).reshape(-1,1)
Y=np.array(Y).reshape(-1,1)
import matplotlib.pyplot as plt

plt.scatter(X,Y)
```

```python

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=0)
```

```python
X_train=np.array(X_train).reshape(-1,1)
Y_train=np.array(Y_train).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
lr.fit(X_train,Y_train)
```

```python
y_pred_train=lr.predict(X_train)
```

```python
plt.scatter(X_train,y_pred_train,color="red")
plt.scatter(X_train,Y_train)
```

```python
y_pred_test=lr.predict(X_test)
```

```python
plt.scatter(X_test,y_pred_test,color="red")
plt.scatter(X_test,Y_test)
```

```python

y_pred_train=np.array(y_pred_train).reshape(-1,1)
y_pred_test=np.array(y_pred_test).reshape(-1,1)
y_pred_test_resize=np.resize(y_pred_test,(90,1))
l=y_pred_train.shape
print(l)
m=y_pred_test_resize.shape
print(m)
```

```python
y_pred=np.concatenate((y_pred_train,y_pred_test_resize),axis=0)
```

```python
print(y_pred.shape)
```

```python
y_pred=np.resize(y_pred,(150,1))
```

```python
print(y_pred.shape)
```

```python
from sklearn.metrics import mean_absolute_error

MAE=mean_absolute_error(Y,y_pred)
print(f"MAE: {MAE}")
```

```python
plt.scatter(X,Y)
plt.scatter(X,y_pred, color="red")
```
